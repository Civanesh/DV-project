```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
dt = pd.read_csv('dataset.csv')
```

```python
sns.heatmap(dt.isnull())
```

```python
dt.info()
```

```python
dt['type'] = dt['type'].fillna(dt['type'].mode()[0])
```

```python
pd.set_option('display.max_rows',500)
```

```python
dt['electors'] = dt['electors'].str.replace(',',"")
dt['electors'] = dt['electors'].str.replace('-',"0")
dt['electors'] = dt['electors'].astype('float',errors='ignore')
```

```python
dt['votes'] = dt['votes'].str.replace(',',"")
dt['votes'] = dt['votes'].str.replace('-',"")
dt['votes'] = pd.to_numeric(dt['votes'],errors='coerce')
```

```python
dt['Turnout'] = dt['Turnout'].str.replace('%',"")
dt['Turnout'] = dt['Turnout'].str.replace('-',"")
dt['Turnout'] = pd.to_numeric(dt['Turnout'],errors='coerce')
```

```python
dt['margin'] = dt['margin'].str.replace(',',"")
dt['margin'] = dt['margin'].str.replace('-',"")
dt['margin'] = pd.to_numeric(dt['margin'],errors='coerce')
```

```python
dt['margin%'] = dt['margin%'].str.replace('%',"")
dt['margin%'] = dt['margin%'].str.replace('-',"")
dt['margin%'] = pd.to_numeric(dt['margin%'],errors='coerce')
```

```python
dt.head()
```

```python
dt.info()
```

```python
dt['electors'] = pd.to_numeric(dt['electors'],errors='coerce')
```

```python
dt.head()
```

```python
dt['electors'] = dt['electors'].fillna(dt['electors'].mean())
dt['votes'] = dt['votes'].fillna(dt['votes'].mean())
dt['Turnout'] = dt['Turnout'].fillna(dt['Turnout'].mean())
dt['margin'] = dt['margin'].fillna(dt['margin'].mean())
dt['margin%'] = dt['margin%'].fillna(dt['margin%'].mean())
```
